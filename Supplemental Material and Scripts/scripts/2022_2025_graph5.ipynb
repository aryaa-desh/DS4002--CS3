```python
import pandas as pd

df = pd.read_csv('/content/preprocessed_data.csv')
display(df.head())
```

# Task
Load the file `preprocessed_data.csv` into a dataframe and create a line plot for the United States for the years 2022-2025.

## Data wrangling

### Subtask:
Filter the data to include only the 'United States' row and the columns for the years 2022 to 2025, plus the 'State' column.


**Reasoning**:
Filter the dataframe to include only the 'United States' row and the specified year columns.



```python
df_filtered = df[df['State'] == 'United States'][['State', '2022', '2023', '2024', '2025']]
display(df_filtered)
```

## Data transformation

### Subtask:
Reshape the filtered data from wide to long format to prepare it for plotting.


**Reasoning**:
Reshape the filtered dataframe from wide to long format using the melt method.



```python
df_long = df_filtered.melt(id_vars=['State'], var_name='Year', value_name='Price of Electricity (cents per kilowatthour)')
display(df_long)
```

## Data visualization

### Subtask:
Create a line plot of the data, showing the trend for the United States from 2022 to 2025.


**Reasoning**:
Create a line plot using the melted dataframe to visualize the trend for the United States from 2022 to 2025 with appropriate labels and title.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df_long['Year'], df_long['Price of Electricity (cents per kilowatthour)'])
plt.title('United States Average Price of Electricity Per Year (2022-2025)')
plt.xlabel('Year')
plt.ylabel('Price of Electricity (cents per kilowatthour)')
plt.show()
```
